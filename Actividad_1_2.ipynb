```python
df= pd.read_excel('DatosFacturación.xlsx')
df
```

```python
# Filtro 1.2
# Columna CVE_CLPV (valores: 1000 a 2000)
filtro1_2=df[(df["CVE_CLPV"] >= 1000) & (df["CVE_CLPV"] <= 2000)]
filtro1_2.to_csv("Filtro1_2.csv")
from google.colab import files 
files.download("Filtro1_2.csv")
filtro1_2
```

```python
# Filtro 2.2
# Columna CVE_VEND (Todas las claves excepto 5 y 4)
filtro2_2=df[~(df["CVE_VEND"] == 5) & ~(df["CVE_VEND"] == 4)]

filtro2_2.to_csv("Filtro2_2.csv")
from google.colab import files 
  
files.download("Filtro2_2.csv")

filtro2_2
```

```python
# Filtro 4.2
# Columna CAN_TOT (Cantidades menores a 5951.7) o STATUS('E')
filtro4_2=df[(df["CAN_TOT"] < 5951.7) | (df["STATUS"] == 'E')]

filtro4_2.to_csv("Filtro4_2.csv")
from google.colab import files 
  
files.download("Filtro4_2.csv")

filtro4_2
```

```python
# Filtro 5.2
# Solo las columnas: CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT
filtro5_2=df.loc[ : , ['CVE_DOC','FECHA_ENT','FECHA_VEN','CAN_TOT']]
filtro5_2.to_csv("Filtro5_2.csv")
from google.colab import files 
files.download("Filtro5_2.csv")
filtro5_2
```

```python
# Filtro 6.2
# Solo las filas de 7001-7099
filtro6_2= df.iloc[7001:7100,: ]
filtro6_2.to_csv("Filtro6_2.csv")
from google.colab import files 
files.download("Filtro6_2.csv")
filtro6_2
```

```python
# Filtro 7.2
# Index = CVE_VEND (valores:1,2) (columna:FECHAELAB)
df2= pd.read_excel('DatosFacturación.xlsx', index_col='CVE_VEND')
filtro7_2=df2.loc[[1,2],['FECHAELAB']]
filtro7_2.to_csv("Filtro7_2.csv")
from google.colab import files 
files.download("Filtro7_2.csv")
filtro7_2
```
