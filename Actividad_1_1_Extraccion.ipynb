**FILTROS BASICOS PARA VARIABLES CUALITATIVAS**

```python
#Filtro por objeto
#Filtrar por la columna NOMBRE_VENDEDOR aquellas filas que contenga el objeto LETICIA RAMIREZ HERNANDEZ
filtro1=df[df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ"]

filtro1.to_csv("Filtro1.csv")
from google.colab import files 
  
files.download("Filtro1.csv")

filtro1
```

```python
#Filtro por filas
#Filtrar las filas no. 57, 342, 871
filtro2=df.iloc[ [57, 342, 871] , : ]  #Filas no consecutivas 

filtro2.to_csv("Filtro2.csv")
from google.colab import files 
  
files.download("Filtro2.csv")

filtro2
```

```python
#Filtro por columnas
#Filtrar a partir de la columna no. 3 hasta la no. 6
filtro3=df.iloc[ : , 3:7]       #Columnas consecutivas 

filtro3.to_csv("Filtro3.csv")
from google.colab import files 
  
files.download("Filtro3.csv")

filtro3
```

**FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**

```python
#Filtro por comparacion
#Filtrar aquellos registros que sean mayor a 4000 en la columna COSTO_TOTAL_CALCULADO
filtro6=df[df["COSTO_TOTAL_CALCULADO"] > 4000]

filtro6.to_csv("Filtro6.csv")
from google.colab import files 
  
files.download("Filtro6.csv")

filtro6
```

**FILTROS BASICOS CON CONETORES LOGICOS**

```python
#Filtro Y
#Filtrar aquellos registros que sean menor a 5000 en la columna COSTO_TOTAL_CALCULADO y aquellos que en la columna
#FECHA_DOC contengan el objeto 15-03-2022
filtro7=df[(df["COSTO_TOTAL_CALCULADO"] < 5000) & (df["FECHA_DOC"] == "15-03-2022")]

filtro7.to_csv("Filtro7.csv")
from google.colab import files 
  
files.download("Filtro7.csv")

filtro7
```

```python
#Filtro o
#Filtrar aquellos registros que sean menor a 100 en la columna COSTO_UNITARIO_CALCULADO o aquellos que en la columna
#FECHA_DOC contengan el objeto 21-07-2022
filtro8=df[(df["COSTO_UNITARIO_CALCULADO"] < 100) | (df["FECHA_DOC"] == "2022-07-21")]

filtro8.to_csv("Filtro8.csv")
from google.colab import files 
  
files.download("Filtro8.csv")

filtro8
```

```python
#Filtro not
#Filtrar aquellos registros que NO sean menor a 10000 en la columna MARGEN_TOTAL_CALCULADO y aquellos que en la columna
#FECHA_DOC NO contengan el objeto 08-09-2022
filtro9=df[~(df["MARGEN_TOTAL_CALCULADO"] < 10000) & ~(df["FECHA_DOC"] == "2022-09-08")]

filtro9.to_csv("Filtro9.csv")
from google.colab import files 
  
files.download("Filtro9.csv")

filtro9
```
